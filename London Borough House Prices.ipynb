```python
# Define a fuction that calculates the ratio of average price of 2018 to 2019 
# to see the average increase over two decades
def price_ratio(d):
    y1997 = float(d['avg_price'][d['year']==1997])
    y2017 = float(d['avg_price'][d['year']==2017])
    ratio = [y2017/y1997]
    return ratio

# Check Barnet's avergae price increase between 1998 and 2018
price_ratio(df5[df5['London_Borough']=='Barnet'])

# initilize a dictionary to store price changes by each borough
final = {}
for b in df5['London_Borough'].unique(): 
    borough = df5[df5['London_Borough'] == b]
    final[b] = price_ratio(borough)
print(final) 

# Convert dictionary to pandas data frame
data_ratio = pd.DataFrame(final)
data_ratio.head()

# Transpose data to make wide data as long data
data_ratio_t = data_ratio.T
data_ratio = data_ratio_t.reset_index()
data_ratio.head()

# Rename columns
data_ratio.rename(columns={'index':'London_Borough', 0:'2017 Ratio'}, inplace=True)
data_ratio.head()

# Select top 20 boroughs where house prices incrased greatest.
top20 = data_ratio.sort_values(by='2017 Ratio',ascending=False).head(15)
print(top20)

ax = top20[['London_Borough','2017 Ratio']].plot(kind='bar')
ax.set_xticklabels(top20.London_Borough)
```
